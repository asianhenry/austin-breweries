```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
import time
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=True)
```

```python
breweries= pd.read_csv('data/brewery_locations.csv')
```

```python
#get the average rating for all beers for each brewery
rating = []
for url in breweries['untappd_website']:

    browser.visit(url)
    time.sleep(2)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    rating.append(soup.find('span', class_='num').text)
```

```python
#get the number of total ratings across all beers
raters = []
for url in breweries['untappd_website']:

    browser.visit(url)
    time.sleep(2)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    raters.append(soup.find('p', class_='raters').text)
```

```python
breweries['total_rating_count'] = raters
```

```python
breweries['avg_beer_rating'] = rating
```

```python
breweries['total_rating_count']=breweries['total_rating_count'].str.replace('\n','')
breweries['total_rating_count']=breweries['total_rating_count'].str.replace(',','')
breweries['total_rating_count']=breweries['total_rating_count'].str.replace(' Ratings','')
breweries['total_rating_count']=pd.to_numeric(breweries['total_rating_count'],errors='coerce')
```

```python
#clean avg rating column

breweries['avg_beer_rating']=breweries['avg_beer_rating'].str.replace('(','')
breweries['avg_beer_rating']=breweries['avg_beer_rating'].str.replace(')','')
breweries['avg_beer_rating']=pd.to_numeric(breweries['avg_beer_rating'],errors='coerce')
```
