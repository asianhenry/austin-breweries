```python
import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from config import gkey
```

```python
breweries= pd.read_csv('data/breweries.csv')
```

```python
breweries.head()
```

```python
address_zip=[]
for i in range(len(breweries)):
    name = breweries['name'][i]
    zip = str(breweries['zipcode'][i])
    addzip = f'{name} {zip}'
    address_zip.append(addzip)
    addzip = ''
#str(zipcode)       
```

```python
#address_zip
```

```python
address = address_zip[0]
query_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(address, gkey)
```

```python
# Run a request to endpoint and convert result to json
geo_data = requests.get(query_url).json()
```

```python
#print(json.dumps(geo_data, indent=4, sort_keys=True))
```

```python
place_id = geo_data['results'][0]['place_id']
```

```python
name = []
lat = []
lng = []
google_address = []
google_rating = []
for i in address_zip:
    place_url =(f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={i}&inputtype=textquery&fields=formatted_address,name,rating,opening_hours,geometry&key={gkey}')
    place_data = requests.get(place_url).json()
    try:
        name.append(place_data['candidates'][0]['name'])
        lat.append(place_data['candidates'][0]['geometry']['location']['lat'])
        lng.append(place_data['candidates'][0]['geometry']['location']['lng'])
        google_address.append(place_data['candidates'][0]['formatted_address'])
        google_rating.append(place_data['candidates'][0]['rating'])
    except:
        name.append('n/a')
        lat.append('n/a')
        lng.append('n/a')
        google_address.append('n/a')
        google_rating.append('n/a')
```

```python
place_url =(f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={name[6]}&inputtype=textquery&fields=formatted_address,name,rating,opening_hours,geometry&key={gkey}')
place_data = requests.get(place_url).json()
```
